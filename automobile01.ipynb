```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Automobile_data.csv")
print(df.info())
df.head()
```

```python
## what is the average price and average mileage for convertible cars
```

```python
df_s1 = df[df["body-style"]=="convertible"]
df_s1
```

```python
df_sol1 = pd.DataFrame(df_s1[["price","average-mileage"]].mean(),columns=["Average-Values"]).reset_index()
df_sol1
```

```python
df_sol1.rename(columns={"index":"Variable"},inplace=True)
```

```python
df_sol1
```

```python
## Sort cars by company and price in decending order
```

```python
df_sol2 = df.sort_values(by=["company","price"],ascending=False)
df_sol2.head()
```

```python
##  Save data for 'Toyota' separately
```

```python
df_sol3 = df[df["company"].str.lower()=="toyota"]
df_sol3
```

```python
## which company is producing most expensive car?
```

```python
df_sol4 = df[df["price"].max()==df["price"]][["company","price"]]
df_sol4
```

```python
## Name the cars on basis of maximum length, highest wheelbase, maximum horsepower
```

```python
required_items=["length","wheel-base","horsepower"]
```

```python
df_sol5 = pd.DataFrame()

for i in required_items:
    df_temp = pd.DataFrame(df[df[i]==df[i].max()]["company"]) 
    df_temp["Variable"] = i
    df_sol5 = pd.concat([df_sol5,df_temp])
df_sol5    
```

```python
## company wise average price of cars and order them in decending order
```

```python
df_sol6 = df.groupby(by=["company"],as_index=False)["price"].mean()
df_sol6.sort_values(by=["price"],ascending=False,inplace=True)
df_sol6["price"]=df_sol6["price"].round(2)
df_sol6
```

```python
## create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then "don't buy" else "buy
```

```python
cond = (df["wheel-base"]<100) & (df["length"]<175)
```

```python
df["decision"]=np.where(cond,"don't buy","buy")
df
```

```python
## Select only cars whose decision is "buy"; find average price based on company of car
```

```python
df_sol8 = df[df["decision"]=="buy"][["company","price"]]
df_sol8 = df_sol8.groupby(by=["company"],as_index=False)["price"].mean()
df_sol8
```
